```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FileUpload, Dropdown, Button, VBox, HBox, Output, FloatText
from IPython.display import display, clear_output
from PIL import Image
import io

def display_image(image, title='Image'):
    if image.ndim == 3 and image.shape[-1] == 1:
        image = image.squeeze()
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

def preprocess_image(image_data):
    img = Image.open(io.BytesIO(image_data)).convert('L')
    img = np.array(img)
    img = cv2.resize(img, (224, 224))
    img = img.astype('float32') / 255.0
    img = np.expand_dims(img, axis=-1)
    return img

def apply_kernel(image, kernel):
    processed_image = cv2.filter2D(image, -1, kernel)
    processed_image = np.clip(processed_image, 0, 1)
    return processed_image

sharpening = np.array([[0, -1, 0],
                       [-1, 5, -1],
                       [0, -1, 0]], dtype=np.float32)

box_blur = np.array([[1/9, 1/9, 1/9],
                     [1/9, 1/9, 1/9],
                     [1/9, 1/9, 1/9]], dtype=np.float32)

identity = np.array([[0, 0, 0],
                     [0, 1, 0],
                     [0, 0, 0]], dtype=np.float32)

prewitt_horizontal = np.array([[-1, 0, 1],
                               [-1, 0, 1],
                               [-1, 0, 1]], dtype=np.float32)

prewitt_vertical = np.array([[-1, -1, -1],
                             [ 0,  0,  0],
                             [ 1,  1,  1]], dtype=np.float32)

gaussian_blur = np.array([[1/16, 2/16, 1/16],
                          [2/16, 4/16, 2/16],
                          [1/16, 2/16, 1/16]], dtype=np.float32)

kernels = {
    "Sharpening": sharpening,
    "Box Blur": box_blur,
    "Identity": identity,
    "Prewitt Horizontal": prewitt_horizontal,
    "Prewitt Vertical": prewitt_vertical,
    "Gaussian Blur": gaussian_blur
}

upload = FileUpload(accept='image/*', multiple=False)
kernel_choice = Dropdown(
    options=["template", "custom"],
    value="template",
    description="Kernel:",
)
template_choice = Dropdown(
    options=["Sharpening", "Box Blur", "Identity", "Prewitt Horizontal", "Prewitt Vertical", "Gaussian Blur"],
    value="Sharpening",
    description="Template:",
)
custom_kernel_values = [FloatText(value=0, description=f'Value {i+1}:') for i in range(9)]
custom_kernel_box = VBox(custom_kernel_values)
apply_button = Button(description="Apply Kernel")
output = Output()

def on_upload_change(change):
    clear_output()
    display(upload, kernel_choice)
    for filename, file_info in upload.value.items():
        image_data = file_info['content']
        processed_image = preprocess_image(image_data)
        pixelated_image = cv2.resize(processed_image, (32, 32), interpolation=cv2.INTER_NEAREST)
        pixelated_image = cv2.resize(pixelated_image, (processed_image.shape[1], processed_image.shape[0]), interpolation=cv2.INTER_NEAREST)
        display_image(pixelated_image, title='Pixelated and Grayscaled Image')

        if kernel_choice.value == "template":
            display(template_choice, apply_button, output)
        else:
            display(custom_kernel_box, apply_button, output)

def on_apply_button_clicked(b):
    with output:
        clear_output()
        if kernel_choice.value == "template":
            kernel = kernels[template_choice.value]
        else:
            kernel = np.array([[custom_kernel_values[0].value, custom_kernel_values[1].value, custom_kernel_values[2].value],
                               [custom_kernel_values[3].value, custom_kernel_values[4].value, custom_kernel_values[5].value],
                               [custom_kernel_values[6].value, custom_kernel_values[7].value, custom_kernel_values[8].value]], dtype=np.float32)
        filtered_image = apply_kernel(pixelated_image.squeeze(), kernel)
        display_image(filtered_image, title='Filtered Image')

upload.observe(on_upload_change, names='value')
kernel_choice.observe(on_upload_change, names='value')
apply_button.on_click(on_apply_button_clicked)

display(upload, kernel_choice)
```
